```python
# Create a normal distribution and plot it.

import numpy as np
import matplotlib.pyplot as plt

# Create a normal distribution with mean 0 and standard deviation 1.
mu, sigma = 0, 1
s = np.random.normal(mu, sigma, 1000)

# Plot the distribution.
count, bins, ignored = plt.hist(s, 30, density=True)

# Plot the probability density function.
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

# Show the plot.
plt.show()

```

```python
# Create a simple data set for doing a linear regression, fit a model, and plot it.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create a simple data set.
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)

# Fit the model.
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)

# Predict on new data.
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

# Plot the data and the model prediction.
plt.scatter(x, y)
plt.plot(xfit, yfit);

# Show the plot.
plt.show()
```

```python
# Do the same with plotly

import plotly.express as px
import numpy as np
from sklearn.linear_model import LinearRegression

# Create a simple data set.
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)

# Fit the model.
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)

# Predict on new data.
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

# Plot the data and the model prediction.
fig = px.scatter(x=x, y=y)
fig.add_trace(px.line(x=xfit, y=yfit).data[0])
fig.show()
```

```python
# Create a simple data set for doing a logistic regression, fit a model, and plot it.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Create a simple data set.
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.binomial(1, 1 / (1 + np.exp(-x)))

# Fit the model.
model = LogisticRegression(solver='lbfgs')
model.fit(x[:, np.newaxis], y)

# Predict on new data.
xfit = np.linspace(-2, 6, 1000)
yfit = model.predict_proba(xfit[:, np.newaxis])[:, 1]

# Plot the data and the model prediction.
plt.scatter(x, y)
plt.plot(xfit, yfit);

# Show the plot.
plt.show()

```

```python
# Create a dataset for doing a decision tree, fit a model, then plot the decision tree

import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# Create a simple data set.
rng = np.random.RandomState(0)
x = rng.randn(100, 2)
y = np.logical_xor(x[:, 0] > 0, x[:, 1] > 0)

# Fit the model.
model = DecisionTreeClassifier()
model.fit(x, y)

# Plot the decision tree.
plt.figure(figsize=(10, 10))
plot_tree(model, filled=True);
```

```python
# Create a complex plane and plot it in 3D using plotly

import plotly.graph_objects as go

# Create a complex plane.
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

# Plot the complex plane in 3D.
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.show()
```
